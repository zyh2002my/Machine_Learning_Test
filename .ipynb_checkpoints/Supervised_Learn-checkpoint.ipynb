```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

from sklearn.neighbors import KNeighborsClassifier
#利用训练集对这个分类器进行拟合 KNeighborsClassifier 保存了数据集，以便在预测时计算与邻居之间的距离
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train,y_train)
print("Test set predictions:{}".format(clf.predict(X_test)))
print("Test set accuracy:{:.2f}".format(clf.score(X_test,y_test)))
```

## 随着邻居个数越来越多，决策边界也就越来越平滑，更平滑的边界对应更简单的模型，使用更少的邻居对应更高的模型复杂度

```python

fig ,axes = plt.subplots(1,3,figsize=(10,3))
for n_neighbors, ax in zip([1,3,9],axes):
    # fit方法返回对象本身，所以我们可以将实例化和拟合放在一行代码中
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf,X,fill=True,eps=0.5,ax=ax,alpha=0.4)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title("{}neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_xlabel("feature 1")
axes[0].legend(loc=3)
```

## 模型复杂度和泛化能力之间的关系，在乳腺癌数据集上的研究，数据集分成训练集和测试集，不同邻居个数对训练集和测试集性能进行评估

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=66)
training_accuracy = []
test_accuracy = []
# n_neighbors取值从1到10
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    #构建模型
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # 记录训练集精度
    training_accuracy.append(clf.score(X_train,y_train))
    # 记录泛化精度
    test_accuracy.append(clf.score(X_test,y_test))
plt.plot(neighbors_settings,training_accuracy,label = "training accuracy")
plt.plot(neighbors_settings,test_accuracy,label = "test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

## 上图中最佳性能在邻居数大约6个处，不过最差性能约为88%，结果仍然可以接收

## 回归的K近邻算法在scikit-learn 的KNeighborsRegressor类中实现

```python
from sklearn.neighbors import KNeighborsRegressor
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train,y_train)
print("Test set predictions:{}".format(reg.predict(X_test)))

# 对于回归问题，返回R2 分数，也叫做决定系数，是回归模型预测的优度度量，1表示完美预测，0对应常熟模型，预测训练集响应y_train的平均值
print("Test set R^2:{:.2f}".format(reg.score(X_test,y_test)))
```

## 分析KNeighborsRegressor

```python
import numpy as np
fig, axes = plt.subplots(1,3,figsize=(15,4))
# 创建1000个数据点，在-3和3之间均匀分布
# 在指定的间隔内返回均匀间隔的数字。
# reshape函数是针对对目标函数取数据用于重新布局时，是按照列的方式来获取数据的
# 方式一：arr.reshape((m,n))
# 意思是把arr矩阵变成一个新的m行n列的矩阵

# 方式二：arr.reshape(-1,1)
# -1表示按照行的方式的获取，结果形成一列

# 方式三:arr.reshape(m,n,q)
# 得到多维数组
line = np.linspace(-3,3,1000).reshape(-1,1)
# print(line)
# for n_neighbors, ax in zip(n_neighbors=n_neighbors)
print(zip([1,3,9],axes))
for n_neighbors, ax in zip([1,3,9],axes):
    # 利用1个，3个或者9个邻居分别进行预测
    print(n_neighbors,ax)
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line,reg.predict(line))
    ax.plot(X_train,y_train,'^',c=mglearn.cm2(0),markersize=8)
    ax.plot(X_test,y_test,'v',c=mglearn.cm2(1),markersize=8)
    # 'numpy.float64 使用{:.2f}
    ax.set_title("{} neighbor(s)\n train score:{:.2f} test score:{:.2f}".format(n_neighbors,reg.score(X_train,y_train),reg.score(X_test,y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions","Trainimng data/target","Test data/target"],loc="best")

```

## KNeighbors 分类器2个重要参数，邻居个数和数据点之间的距离度量方法，实践中，一般使用较小的邻居个数比如3，5个可以得到很好的效果。度量方法一般使用欧式距离
## k近邻优点是很容易理解，缺点是预测速度慢，且不能处理具有很多特征的数据集，所以一般不常用。线性模型没有这个两个缺点

## 线性模型 利用输入特征的线性函数进行预测 预测公式如下 y = w[0]*x[0]+b 
## 其中 x为特征，y是模型的预测结果，w和b是学习模型的参数

```python
mglearn.plots.plot_linear_regression_wave()
```

## 线性回归，一个特征是一条直线，2个是一个面，多个是更高纬度的超平面，特征数大于训练数据点数量，任何目标y可以用线性函数完美拟合，有许多的线性回归模型，区别在于如何从训练数据中学习参数w和b，以及如何控制模型复杂度

## 线性回归 又名普通最小二乘法  寻找参数w和b，使得对训练集的预测值与真实的回归目标值y之间均方误差最小。均方误差是 预测值与真实值之差的平方和除以样本数

```python
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train,y_test = train_test_split(X,y, random_state=42)
lr = LinearRegression().fit(X_train,y_train)
print("lr.coef_:{}".format(lr.coef_))
print("lr.intercept_:{}".format(lr.intercept_))

print("Trainimng set score:{:.2f}".format(lr.score(X_train,y_train)))
print("Test set score:{:.2f}".format(lr.score(X_test,y_test)))
```

## 训练集和测试集上分数接近，存在欠拟合，对于更高维的数据集，拥有大量特征，过拟合可能会变大

```python
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)

print("Trainimng set score:{:.2f}".format(lr.score(X_train, y_train)))
print("Test set score:{:.2f}".format(lr.score(X_test,y_test)))
```

## 训练集和测试集之间的性能差异是过拟合的明显标志

## 岭回归代替线性回归，对系数 w 的所有元素都应该接近于0，意味着每个特征对输出的影响尽可能小，即斜率小对w拟合附加约束，这种约束叫正规化 regularization。

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print("Trainimng set score:{:.2f}".format(ridge.score(X_train, y_train)))
print("Test set score:{:.2f}".format(ridge.score(X_test,y_test)))
```

## Ridge 是一种约束更强的模型，所以更不容易过拟合，复杂度更小的模型意味着在训练集上性能更差，但泛化性能更好，增大alpha会使得系数更加趋向于0

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Trainimng set score:{:.2f}".format(ridge10.score(X_train, y_train)))
print("Test set score:{:.2f}".format(ridge10.score(X_test,y_test)))
```

## 减小alpha可以让系数受到的限制更小，对于非常小的alpha值，系数几乎没有收到限制

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Trainimng set score:{:.2f}".format(ridge01.score(X_train, y_train)))
print("Test set score:{:.2f}".format(ridge01.score(X_test,y_test)))
```

## 更大的alpha表示约束更强的模型，预计大的alpha对应的coef_元素比小alpha对应的coef_元素要小

```python
plt.plot(ridge.coef_,'s',label="Ridge alpha=1")
plt.plot(ridge10.coef_,'^',label="Ridge alpha=10")
plt.plot(ridge01.coef_,'v',label="Ridge alpha=0.1")
plt.xlabel("Coefficient index")
plt.ylabel("Coeffici magnitude")

plt.plot(lr.coef_,'o',label="LinearRegression")
plt.hlines(0,0,len(lr.coef_))
plt.ylim(-25,25)
plt.legend()
```

## Lasso 另一种正规化线性回归，使用L1 正规化，某些系数刚好为0，自动化特征选择，模型更容易理解，也可以呈现最重要的特真

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print("Trainimng set score:{:.2f}".format(lasso.score(X_train, y_train)))
print("Test set score:{:.2f}".format(lasso.score(X_test,y_test)))
```

```python
from sklearn.linear_model import Lasso

lasso001 = Lasso(alpha=0.01,max_iter=100000).fit(X_train, y_train)
print("Trainimng set score:{:.2f}".format(lasso001.score(X_train, y_train)))
print("Test set score:{:.2f}".format(lasso001.score(X_test,y_test)))
print("Number of features used:{}".format(np.sum(lasso001.coef_!=0)))
```

```python
from sklearn.linear_model import Lasso

lasso00001 = Lasso(alpha=0.0001,max_iter=100000).fit(X_train, y_train)
print("Trainimng set score:{:.2f}".format(lasso00001.score(X_train, y_train)))
print("Test set score:{:.2f}".format(lasso00001.score(X_test,y_test)))
print("Number of features used:{}".format(np.sum(lasso00001.coef_!=0)))
```

## alpha值变小 我们可以拟合一个更复杂的模型，在训练集和测试集上表现也更好，alpha如果太小，就会消除正则化效果，并出现过拟合，得到与LinearRegressin的类似效果

## 分类的线性模型，线性模型也广泛应用于分类问题
## 对于用于回归的线性模型，输出y是特征的线性函数，对于分类的线性模型，决策边界是输入的线性模型
## 学习线性模型有很多算法1.系数和偏移的特定组合对训练数据拟合好坏的度量方法2.是否使用正则化，以及使用哪种正则化方法


## 常见的两种线性分类算法是Logistic回归（logistic regression）和线性支持向量机（linear support vector machine 线性svm）

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X, y = mglearn.datasets.make_forge()
fig, axes = plt.subplots(1, 2, figsize=(10,3))
# print(axes[:1,])
for model, ax in zip([LinearSVC(),LogisticRegression()],axes):
#     print(model)
#     print(ax)
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5,ax=ax,alpha=.7)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title("{}".format(clf.__class__.__name__))
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
axes[0].legend()
```

## 决定正则化强度的权衡参数叫C，C值越大，对应的正则化越弱，C值较小，模型更强调使系数向量（w）接近于0，较小的C值可以让算法尽量适应“大多数”数据点，较大的C值更强调每个数据点都分类正确的重要性

## C值小，对应强正则化，模型会选择一条相对水平的线，最右边的模型尽量使所有的点分类都正确，但可能无法掌握类别的整体分布，这个模型可能会过拟合

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_train, X_test, y_train,y_test = train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=42)
logreg = LogisticRegression().fit(X_train,y_train)


print("Training set score:{:.3f}".format(logreg.score(X_train,y_train)))
print("Test set score:{:.3f}".format(logreg.score(X_test,y_test)))
```

## 训练集和测试集性能非常接近，模型很可能是欠拟合

```python
logreg100 = LogisticRegression(C=100).fit(X_train,y_train)


print("Training set score:{:.3f}".format(logreg100.score(X_train,y_train)))
print("Test set score:{:.3f}".format(logreg100.score(X_test,y_test)))
```

```python
logreg001 = LogisticRegression(C=0.01).fit(X_train,y_train)


print("Training set score:{:.3f}".format(logreg001.score(X_train,y_train)))
print("Test set score:{:.3f}".format(logreg001.score(X_test,y_test)))
```

```python
# 模型系数图
plt.plot(logreg.coef_.T,'o',label="C=1")
plt.plot(logreg100.coef_.T,'^',label="C=100")
plt.plot(logreg001.coef_.T,'v',label="C=0.001")
plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
plt.hlines(0,0,cancer.data.shape[1])
plt.ylim(-5,5)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.legend()
```

## LogisticRegression默认应用L2正规化，和岭回归类似，如果想要一个可解释性更强的模型，使用L1正则化可能更好

```python
for C, marker in zip([0.001,1,100],['o','^','v']):
    lr_l1 = LogisticRegression(C=C,penalty="l1").fit(X_train,y_train)
   # print(C,type(lr_l1.score(X_train, y_train)))
    print("Training accuracy of l1 logreg with C={:.3f}:{:.2f}".format(C,lr_l1.score(X_train, y_train)))
    print("Test accuracy of l1 logreg with C={:.3f}:{:.2f}".format(C,lr_l1.score(X_test, y_test)))
    
    plt.plot(lr_l1.coef_.T, marker, label="C={:.3f}".format(C))
    plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
    plt.hlines(0,0,cancer.data.shape[1])
    plt.ylim(-5,5)
    plt.xlabel("Coefficient index")
    plt.ylabel("Coefficient magnitude")
    plt.legend(loc=3)
```

## 用于而分类的线性模型与用于回归的线性模型有许多相似之处，与用于回归的线性模型一样，模型主要差别在于penalty参数，影响正则化，模型是使用所有的可用特征还是只选择一个子集

## 许多线性分类模型只适用于二分类问题，不能轻易推广到多类别问题 除了Logistic回归，其中一种常见方法是“一对其余” one-vs-rest方法，每个类别都学习一个二分类模型，

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(random_state=42)
#print(X,y)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0","Class 1","Class 2"])
```

```python
linear_svm = LinearSVC().fit(X,y)
print("Coefficient shape:",linear_svm.coef_.shape)
print("Intercept shape :",linear_svm.intercept_.shape)
```

## coef_每行包含三个类别之一的系数向量，每列包含2个特种，intercept_是一维数组，保存每个类别的截距

```python
mglearn.discrete_scatter(X[:,0],X[:,1],y)
line = np.linspace(-15,15)
#print(line)
for coef,intercept,color in zip(linear_svm.coef_,linear_svm.intercept_,['b','r','g']):
    print(coef)
    plt.plot(line,-(line*coef[0]+intercept)/coef[1],c=color)
plt.ylim(-10,15)
plt.xlim(-10,8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(['Class 0','Class 1','Class 2','Line class 0','Line class 1','Line class 2'],loc=(1.01,0.3))
    
```

## 三个“一对其余”分类器学到的决策边界

```python
mglearn.plots.plot_2d_classification(linear_svm,X,fill=True,alpha=.7)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
line = np.linspace(-15,15)
# #print(line)
for coef,intercept,color in zip(linear_svm.coef_,linear_svm.intercept_,['b','r','g']):
#     # k=（y1-y2）/（x1-x2）
    plt.plot(line,-(line*coef[0]+intercept)/coef[1],c=color)
# plt.ylim(-10,15)
# plt.xlim(-10,8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1") 
plt.legend(['Class 0','Class 1','Class 2','Line class 0','Line class 1','Line class 2'],loc=(1.01,0.3))
```

```python
X = np.array([[0,1,0,1],
              [1,0,1,1],
              [0,0,0,1],
              [1,0,1,0]])
y = np.array([0,1,0,1])

counts = {}
for label in np.unique(y):
    # 对每个类别进行遍历
    # 计算（求和）每个特征中1的个数
#     print(label)
#     print(y == label)
#     print(X[y == label])
    counts[label] = X[y == label].sum(axis=0)
print("Feature counts:\n{}".format(counts))

#print(X[np.array([0,1,0,1])])
```

## 构造决策树所有叶节点都是纯的叶节点，会导致模型特别复杂，过拟合，防止过拟合有2种常见的策略，预剪枝pre-pruning 后剪枝 post-pruning 预剪枝限制树的最大深度，叶节点的最大数目，或者一个节点种的最小数据点数目来防止划分

## sklearn只实现了pre-pruning ，未实现后剪枝，未剪枝的树容易过拟合，对新的数据泛化性能不佳

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

## 树不能在训练数据的范围之外生成新的响应，所有基于树的模型都有这个缺点

## 决策树有点是 1 模型容易理解和可视化 ，2 特征不需要预处理，尺度不一样时，二元特征和连续特征同时存在时，效果好
## 缺点是使用了预剪枝，也会过拟合，泛化性能差，所以一般使用决策树集成 ensemble ，
## 常用两类 都是以决策树作为基础，1 是随机森林 random forest 2 是梯度提升决策树gradient boosted decision tree

##  随机森林 构造时 ，2种方法 1.通过选择用于构造树的数据点， 2是通过选择每次划分的测试特征。
## 比如我们要构造10棵树，构造时完全独立，对数据进行自助采样 bootstrap sample，会有三分之一的数据丢失，从n_samples个数据点中，重复随机抽取一个样本，抽取n_samples次
## 下面构造决策树，每个节点处，算法随机选择特征的一个子集，对一个特征寻找最佳测试 特征个数由max_features来控制
## max_features较大， 树会非常相似，利用最独特的特征可以轻松拟合数据，max_features较小，树差异会很大，为了很少的拟合数据，每个树深度都会很大

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                    random_state=42)
forest = RandomForestClassifier(n_estimators = 5,random_state=2)
forest.fit(X_train, y_train)
```

```python
fig, axes = plt.subplots(2,3,figsize=(20,10))
for i,(ax,tree) in enumerate(zip(axes.ravel(),forest.estimators_)):
    #print(i,ax,tree)
    ax.set_title("Tree{}".format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train,tree,ax=ax)
mglearn.plots.plot_2d_separator(forest, X_train, fill=True, ax=axes[-1, -1],
                                alpha=.4)
# 确定坐标范围    
axes[-1,-1].set_title("Random Forest")
#print(axes)
#print(type(axes))
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
    
```

## 随机森林比单独每一棵树的过拟合都要小，在实际应用中，我们会用更多棵树的 几百或上千，得到更平滑的边界

```python
from sklearn.ensemble import GradientBoostingClassifier
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=0)
gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train,y_train)
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

## 如果将线性SVM模型看作原始特征函数，那么它实际上已经不是线性的了，它不是一条直线，而是一个椭圆
## 向数据表示中添加非线性特征，可以让线性模型变得更强大，

## 通常只有一部分训练数据点对于定义决策边界很重要 位于类别之间边界上的那些点，叫做支持向量，支持向量机正式由此得名，想要对新样本点进行预测，需要测试它与每个支持向量之间的距离，分类决策是基于它与支持向量之间的距离以及在训练过程中学到的支持向量重要性（保存在SVC的dual_coef_属性中）来做出的
## 数据点之间的距离由高斯核给出：

```python
from sklearn.svm import SVC
X, y = mglearn.tools.make_handcrafted_dataset()
svm = SVC(kernel='rbf',C=10,gamma=0.1).fit(X,y)
mglearn.plots.plot_2d_separator(svm,X,eps=.5)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
# 画出支持向量
sv = svm.support_vectors_
# 支持向量的类别标签由dual_coef_的正负号给出
print(svm.dual_coef_)
#ravel()函数可以去除ndarray格式直接表示元素
print(svm.dual_coef_.ravel())
sv_labels = svm.dual_coef_.ravel()>0
mglearn.discrete_scatter(sv[:,0],sv[:,1],sv_labels,s=15,markeredgewidth=3)
```

## svm给出了非常平滑且非线性不是直线的边界，我们调节了C和gamma参数
## gamma 是控制高斯核的宽度，决定了点与点之间“靠近”多大的距离，C是正规化参数，与线性模型中用到的类似，它限制每个点的重要性，每个点的 dual_coef_

```python
fig, axes = plt.subplots(3,3,figsize=(15,10))
for ax,C in zip(axes,[-1,0,3]):
    for a,gamma in zip(ax,range(-1,2)):
        mglearn.plots.plot_svm(log_C=C,log_gamma=gamma,ax=a)
axes[0, 0].legend(["class 0", "class 1", "sv class 0", "sv class 1"],
                  ncol=4, loc=(.9, 1.2))
```

## gamma小 高斯核半径大，许多点看作比较近，越右决策边界更关注单个点，小的gamma值表示决策边界变化很慢，生成的是复杂度较低的模型
## 与线性模型形同，C值很小，说明模型非常受限，每个数据点影响范围有限

## 过拟合了，svm对参数的设定和数据的缩放非常敏感，它要求所有特征有相似的变化范围，看一下特征的最大和最小值

```python
plt.plot(X_train.min(axis=0),'o',label="min")
plt.plot(X_train.max(axis=0),'^',label="max")
plt.yscale("log")
# print(X_train)
# print(X_train.min(axis=0))
# #axis None代表所有的最小值，1代表每一行的最小值，0代表每一列的最小值
# arr1=np.array([[3,1,2],[5,4,6],[7,8,0]])
# print(arr1.min(axis=0))
plt.legend(loc=4)
plt.xlabel("Feature index")
plt.ylabel("Feature magnitude")
```

```python
# 为svm预处理数据，对每个特征进行缩放，使其大致都位于同一范围，核SVM常用的缩放方法就是将所有特征缩放到0和1之间
# 计算训练集中每个特征的最小值
min_on_training = X_train.min(axis=0)
#print(min_on_training)
# 计算训练集中每个特征的范围（最大-最小）
range_on_training = (X_train - min_on_training).max(axis=0)
# print(range_on_training)
# print("\n{}".format(X_train - min_on_training))
# 减去最小值 然后除以范围
# 这样每个特征都是min=0和max=1
X_train_scaled = (X_train - min_on_training)/range_on_training
print("Minimum for each feature \n{}".format(X_train_scaled.min(axis=0)))
print("Maximum for each feature \n{}".format(X_train_scaled.max(axis=0)))
```

```python
# 利用训练集的最小值和范围对测试集做相同的变换
X_test_scaled = (X_test - min_on_training)/range_on_training
```

```python
svc = SVC()
svc.fit(X_train_scaled,y_train)
print("Accuracy on training set: {:.3f}".format(svc.score(X_train_scaled, y_train)))
print("Accuracy on test set: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

## 现在的模型现在处于欠拟合的状态，因为训练集和测试集的性能非常接近，还没有接近100%的精度，可是使用增大C或gamma来拟合更为复杂的模型


```python
svc = SVC(C=1000)
svc.fit(X_train_scaled,y_train)
print("Accuracy on training set: {:.3f}".format(svc.score(X_train_scaled, y_train)))
print("Accuracy on test set: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

## svm 允许决策边界很复杂 即使数据只有几个特征，它在低纬度和高纬度上的表现都很好，但对样本个数的缩放表现不好
## svm的另一个缺点是 预处理数据和调参都需要小心 这也是为什么如今很多的应用都是基于树模型，比如随机森林和梯度提升，他们不需要预处理，另外svm模型很难检查，很难理解为什么这么预测
## svm特别试用于所有特征的测量单位类似，比如都是像素密度，而且范围差不多
## 核svm重要参数是正规化参数C和gamma，都是控制模型复杂度，较大的值都对应更为复杂的模型，

## 在MLP中，多次重复这个计算加权求和的过程，首先计算代表中间过程的隐单元hidden unit，然后再计算这些隐单元的加权求和并得到最终的结果



```python
line = np.linspace(-3,3,100)
plt.plot(line,np.tanh(line),label='tanh')
plt.plot(line,np.maximum(line,0),label='relu')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('relu(x),tanh(x)')
```

## 有两个隐层的多层感知机 这些由许多计算层组成的大型神经网络，正式术语深度学习的来源

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.25,random_state=3)
X_train, X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=42)

mlp = MLPClassifier(solver='lbfgs',random_state=0).fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

## 神经网络学到的决策边界完全是非线性的，相对平滑，MLP使用100个隐节点，对于这个小型数据集来说已经很多了，减少数量，降低模型复杂度试试，也有很好的结果


```python
mlp = MLPClassifier(solver='lbfgs',random_state=0,hidden_layer_sizes=[10])
mlp.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

## 只有10个隐单元 ，决策边界看起来更加参差不齐，默认的非线性是relu，如果使用单隐层，那么决策函数由10个直线组成，想得到更平滑的决策边界，可以添加更多的隐单元，添加第二个隐层或者使用tanh非线性

```python
# 使用2个隐层，每个包含10个单元
mlp = MLPClassifier(solver='lbfgs',random_state=0,hidden_layer_sizes=[10,10])
mlp.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
# 使用2个隐层，每个包含10个单元，这次使用tanh非线性
mlp = MLPClassifier(solver='lbfgs',activation = 'tanh',random_state=0,hidden_layer_sizes=[10,10])
mlp.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
